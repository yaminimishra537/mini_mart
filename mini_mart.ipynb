```python
import time
import random
from IPython.display import display, Image

#  GIFS
def show_gif(url):
    display(Image(url=url))

welcome_gif = "https://media2.giphy.com/media/v1.Y2lkPTc5MGI3NjExNm4ydzJsZ2NidWF2M2QxZmpxY2YyNHVuMWh4ZmNsanZsM29vcXVzbyZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/WUOtxse2uwDi2CACWM/giphy.gif"
add_gif = "https://media1.giphy.com/media/v1.Y2lkPTc5MGI3NjExdXY3eTI4dWhxdWVjY2YwcHg3bjZla3BmajZkaWN1eHFsemJ1cWhkbiZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/43OKja0fROZxpFoQeg/giphy.gif"
checkout_gif = "https://media0.giphy.com/media/v1.Y2lkPTc5MGI3NjExOWFsbTA2MTdnZndzM3ZlYXhqMnJqbzdmMjF3a3Fnd2doa2ZzbjNleiZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/1BPvv1QvbKV156jyul/giphy.gif"
game_gif = "https://media3.giphy.com/media/v1.Y2lkPTc5MGI3NjExdjZzbWI3a2QyZ3FoOGh2ZzRsc21ldTNmOWppOWVpaDl6MXV6ZHNldiZlcD12MV9pbnRlcm5hbF9naWZfYnlfaWQmY3Q9Zw/Rs2QPsshsFI9zeT4Kn/giphy.gif"

#  USER LOGIN
username = input("Enter your name: ")
mobile = input("Enter your mobile number: ")
password = input("Enter your password: ")
location = input("Enter your location: ")
print(f" Welcome {username}! You get 10% discount on Ganesh Chaturthi & free delivery on first 2 orders.")
show_gif(welcome_gif)

#  CART
cart = []
cart_prices = []

points = 0  # points from game for discount

#  SECTIONS
def food_section():
    print("\n--- FOOD SECTION ---")
    print("1. Pizza - 150")
    print("2. Burger - 100")
    print("3. Ice Cream - 50")
    print("4. Coke - 50")
    print("5. Combo: Pizza + Coke - 180")
    print("6. Combo: Burger + Ice Cream - 130")
    choice = input("Enter item name to add: ")

    if choice.lower() == "pizza":
        cart.append("Pizza")
        cart_prices.append(150)
    elif choice.lower() == "burger":
        cart.append("Burger")
        cart_prices.append(100)
    elif choice.lower() == "ice cream":
        cart.append("Ice Cream")
        cart_prices.append(50)
    elif choice.lower() == "coke":
        cart.append("Coke")
        cart_prices.append(50)
    elif choice.lower() == "combo1" or choice.lower() == "pizza + coke":
        cart.append("Pizza + Coke Combo")
        cart_prices.append(180)
    elif choice.lower() == "combo2" or choice.lower() == "burger + ice cream":
        cart.append("Burger + Ice Cream Combo")
        cart_prices.append(130)
    else:
        print(" Item not available")
        return
    print(f" {choice} added to cart")
    show_gif(add_gif)

def clothes_section():
    print("\n--- CLOTHES SECTION ---")
    print("1. T-Shirt - 800")
    print("2. Jeans - 1500")
    print("3. Kurti - 199")
    choice = input("Enter item name to add: ")

    if choice.lower() == "t-shirt":
        cart.append("T-Shirt")
        cart_prices.append(800)
        print(" Recommendation: If you buy T-Shirt, you may also like Jeans!")
    elif choice.lower() == "jeans":
        cart.append("Jeans")
        cart_prices.append(1500)
    elif choice.lower() == "kurti":
        cart.append("Kurti")
        cart_prices.append(199)
    else:
        print(" Item not available")
        return
    print(f" {choice} added to cart")
    show_gif(add_gif)

def grocery_section():
    print("\n--- GROCERY SECTION ---")
    print("1. Biscuits - 50 (Buy 1 Get 2)")
    print("2. Cake - 200 (Buy 1 Get 2)")
    print("3. Frozen Peas - 80 (Discount 20%)")
    print("4. Baby Powder - 250 (Discount 15%)")
    choice = input("Enter item name to add: ")

    if choice.lower() == "biscuits":
        cart.append("Biscuits x3")
        cart_prices.append(50)
    elif choice.lower() == "cake":
        cart.append("Cake x3")
        cart_prices.append(200)
    elif choice.lower() == "frozen peas":
        cart.append("Frozen Peas")
        cart_prices.append(80*0.8)
    elif choice.lower() == "baby powder":
        cart.append("Baby Powder")
        cart_prices.append(250*0.85)
    else:
        print(" Item not available")
        return
    print(f" {choice} added to cart")
    show_gif(add_gif)

def game_section():
    global points
    print("\n--- GAME SECTION: Guess the Number ---")
    show_gif(game_gif)
    number = random.randint(1,5)
    guess = int(input("Guess a number between 1 and 5: "))
    if guess == number:
        print(" Correct! You get 10% discount on your total purchase")
        points += 10
    else:
        print(f" Wrong! The number was {number}")

def view_cart():
    if not cart:
        print(" Cart is empty")
        return
    print("\n Your Cart:")
    total = 0
    for i in range(len(cart)):
        print(f"{cart[i]} - {cart_prices[i]}")
        total += cart_prices[i]
    print(f" Total: {total}")
    return total

def checkout():
    total = view_cart()
    if points>0:
        discount = total*points/100
        total -= discount
        print(f" Discount from game: {discount}")
    print("\nPayment options: 1. Card  2. Cash  3. UPI")
    pay = input("Choose payment method: ")
    print(" Payment Successful!")
    show_gif(checkout_gif)
    print(f" Final Amount Paid: {total}")
    print(" Thank you for shopping!")
    cart.clear()
    cart_prices.clear()

#  MAIN APP LOOP
while True:
    print("\n--- MAIN MENU ---")
    print("1. Food Section")
    print("2. Clothes Section")
    print("3. Grocery Section")
    print("4. Game Section")
    print("5. View Cart")
    print("6. Checkout & Exit")

    menu_choice = input("Enter choice: ")

    if menu_choice=="1":
        food_section()
    elif menu_choice=="2":
        clothes_section()
    elif menu_choice=="3":
        grocery_section()
    elif menu_choice=="4":
        game_section()
    elif menu_choice=="5":
        view_cart()
    elif menu_choice=="6":
        checkout()
        break
    else:
        print(" Invalid choice")
```
